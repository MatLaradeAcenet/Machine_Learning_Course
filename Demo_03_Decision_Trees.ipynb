# Decision Trees

In this notebook, we go over how to grow decision trees, and look at some of the different paramaters that we can pass to the modules within SciKit-Learn

```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import train_test_split
import numpy as np
iris_attributes, iris_classes = load_iris(return_X_y=True)
attributes_train, attributes_test, classes_train, classes_test = train_test_split(iris_attributes, iris_classes, test_size=0.2, random_state=50)
```

First, similar to naive bays, we create the classifier and assign it to a variable

```python
Tree1 = tree.DecisionTreeClassifier()
```

```python
Tree1  = Tree1.fit(attributes_train, classes_train)
```
